```python
import pandas as pd
import os

dir_name = "../html_files/"

zip_files = os.listdir(dir_name)


def get_zip(fn):
    if fn == "florida_zips":
        return None
    return fn.split("_")[3].split(".")[0]

zip_map = {}

for i in range(0, len(zip_files)):
    zip_map[get_zip(zip_files[i])] = zip_files[i]
    
print(zip_map)
```

```python
def read_file(fn, dirname=dir_name):
    if dirname + fn == '../html_files/florida_zips':
        return None
    f = open(dirname + fn, "r")
    lines = f.read().split("\n")
    return lines

read_file(zip_files[0])

# Store Number
# Distance Away from zip (not wanted)
# Address
# City, State, Zip
# Phone Number
# More Details (unwanted)
# '' Empty String

# add in:
# Zip Code Found Under


all_stores_df = pd.DataFrame()

for key in list(zip_map.keys()):
    zip_ = key
    lines = read_file(zip_map[key])
    data_obj = {
        "Store_Number":[],
        "Address":[],
        "City_State_Zip":[],
        "Phone_Number":[],
        "Zip_Code_Found":[]
    }
    after_space = 0
    if lines is None:
        continue
    else:
        for i in range(0, len(lines)):
            if lines[i] == '':
                data_obj["Zip_Code_Found"] = key
                after_space = 0
            else:
                if after_space == 0:
                    data_obj["Store_Number"].append(lines[i])
                elif after_space == 1:
                    after_space += 1 # continue
                    continue
                elif after_space == 2:
                    data_obj["Address"].append(lines[i])
                elif after_space == 3:
                    data_obj["City_State_Zip"].append(lines[i])
                elif after_space == 4:
                    data_obj["Phone_Number"].append(lines[i])
                after_space += 1
    df = pd.DataFrame(data = data_obj)
    all_stores_df = all_stores_df.append(df, ignore_index=True)
    all_stores_df.to_csv("Nrgize_Store_Locations_Duplication.csv", index=False)
    df.to_csv("Data_for_" + str(key) + ".csv", index=False)
```

```python
print(all_stores_df.head())
print(all_stores_df.shape)

# Store_Number

store_numbers = all_stores_df["Store_Number"]

new_store_numbers = []
for store in store_numbers:
    new_store_numbers.append(store.split(":")[1].strip())
    
all_stores_df["Store_Number"] = new_store_numbers

```

```python
print(all_stores_df)

print(all_stores_df["Store_Number"].value_counts())

# Unique Store_Numbers

unique_stores = all_stores_df["Store_Number"].unique()

# get unique store data:

final_df = pd.DataFrame()

for store in unique_stores:
    df = all_stores_df[all_stores_df["Store_Number"] == store].iloc[0]
    final_df = final_df.append(df)
    
print(final_df.shape)

final_df.to_csv("Nrgize_Store_Locations.csv", index=False)
```
